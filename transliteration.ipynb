```python
import os
import google.generativeai as genai
from datasets import load_dataset

GEMINI_API_KEY = ""
genai.configure(api_key=GEMINI_API_KEY)
ds = load_dataset("SKNahin/bengali-transliteration-data")

generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 40,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
  model_name="gemini-2.0-flash-exp",
  generation_config=generation_config,
  system_instruction="you are transliterating 'banglish' (when bangla is written using english alphabet) into bangla. instead of strictly following a phonetic layout be open so that the transliterated sentences are meaningful in real life.",
)

chat_session = model.start_chat(
  history=[
  ]
)

# text = "vokkor chokkor post akta korlei approved…. nishchoi ghabla ache"

# response = chat_session.send_message(f"Translate the following sentence: '{text}'. Just output the bangla text.")

# print(response.text)
```

```python
# Split the dataset into training and validation subsets (80/20 split)
train_test_split = ds['train'].train_test_split(test_size=0.2)
train_dataset = train_test_split['train']
validation_dataset = train_test_split['test']

print(f"Training dataset size: {len(train_dataset)}")
print(f"Validation dataset size: {len(validation_dataset)}")
```
